Task: Top Cuisines
    
   - Determine the top three most
     common cuisines in the dataset.

   - Calculate the percentage of
     restaurants that serve each of the top
      cuisines.

$Read$ $the$ $necessary$ $packages$

```python
import pandas as pd
import matplotlib.pyplot as plt
```

$Read$ $the$ $Data$

```python
Dataset_df=pd.read_csv(r"C:\Users\ajcho\OneDrive\Desktop\full stack data science and AI\Data Files\Dataset.csv")
Dataset_df
```

```python
Cu_counts=Dataset_df["Cuisines"].value_counts()
Cu_counts
```

```python
top_cu=Cu_counts.head(3)
top_cu
```

```python
total_rest=len(Dataset_df)
total_rest
```

```python
percentage=(top_cu/total_rest)*100
percentage
```

```python
plt.bar(percentage.index,percentage.values)
plt.xlabel('cuisine')
plt.ylabel('Percentage')
plt.title('Top Three Cuisine')
plt.show()
```

```python
plt.pie(top_cu,
        labels=top_cu.index,
        autopct='%0.3f%%',
        explode=[0.1,0.1,0.1])
plt.show()
```
