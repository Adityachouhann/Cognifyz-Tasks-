Task: Restaurant Chains
- Identify if there are any restaurant chains
present in the dataset.
- Analyze the ratings and popularity of
different restaurant chains.

$Read$ $the$ $necessary$ $packages$

```python
import pandas as pd
import matplotlib.pyplot as plt
```

$Read$ $the$ $Data$

```python
Dataset_df=pd.read_csv(r"C:\Users\ajcho\OneDrive\Desktop\full stack data science and AI\Data Files\Dataset.csv")
Dataset_df
```

```python
rest_chains=Dataset_df.groupby('Restaurant Name').size().reset_index(name='Chain Count')
rest_chains=rest_chains[rest_chains['Chain Count']>1]
```

```python
rest_chains=rest_chains.sort_values(by='Chain Count',ascending=False)
```

```python
plt.figure(figsize=(16,12))
plt.bar(rest_chains['Restaurant Name'][:10],rest_chains['Chain Count'][:10])
plt.xlabel('Restaurant chains')
plt.ylabel('Num of Outlets')
plt.title('Top 10 Restaurant Chains by Number of Outlets')
plt.show()
```

```python
chain_ratings=Dataset_df.groupby('Restaurant Name')['Aggregate rating'].mean().reset_index(name='Average Rating')
chain_votes=Dataset_df.groupby('Restaurant Name')['Votes'].sum().reset_index(name='Total Votes')
```

```python
chain_analysis=pd.merge(chain_ratings,chain_votes,on='Restaurant Name')
```

```python
chain_analysis=chain_analysis.sort_values(by='Average Rating',ascending=False)
```

```python
plt.figure(figsize=(16,12))
plt.bar(chain_analysis['Restaurant Name'][:10],chain_analysis['Average Rating'][:10])
plt.xlabel('Restaurant chains')
plt.ylabel('Average Rating')
plt.title('Top 10 Restaurant Chains by Average Rating')
plt.show()
```
