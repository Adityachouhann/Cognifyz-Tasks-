Task: City Analysis
    
   - Identify the city with the highest number
      of restaurants in the dataset.

   - Calculate the average rating for
      restaurants in each city.
      
   - Determine the city with the highest
      average rating.

$Read$ $the$ $necessary$ $packages$

```python
import pandas as pd
import matplotlib.pyplot as plt
```

$Read$ $the$ $Data$

```python
Dataset_df=pd.read_csv(r"C:\Users\ajcho\OneDrive\Desktop\full stack data science and AI\Data Files\Dataset.csv")
Dataset_df
```

```python
rest_count=Dataset_df.groupby('City')['Restaurant ID'].count()
high_rest=rest_count.idxmax()
print(f'City with highest numbers of restaurants is {high_rest}')
```

```python
avg_rating=Dataset_df.groupby('City')['Aggregate rating'].mean()
avg_rating
```

```python
avg_rating.plot(kind='bar',figsize=(28,8))
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Rating for Restaurant in each City')
plt.show()
```

```python
city_with_high_avg_rating=avg_rating.idxmax()
print(f'City with highest average rating :{city_with_high_avg_rating}')
```

```python
print('====================================================================')
print('1=>')

print(f'City with highest numbers of restaurants is {high_rest}')
print('--------------------------------------------------------------------')
print('2=>')

print(avg_rating)
print('--------------------------------------------------------------------')

print('3=>')

print(f'City with highest average rating :{city_with_high_avg_rating}')
print('====================================================================')
```
