Task: Online Delivery
    
   - Determine the percentage of restaurants
      that offer online delivery.
   
   - Compare the average ratings of restaurants
        with and without online delivery.

$Read$ $the$ $necessary$ $packages$

```python
import pandas as pd
import matplotlib.pyplot as plt
```

$Read$ $the$ $Data$

```python
Dataset_df=pd.read_csv(r"C:\Users\ajcho\OneDrive\Desktop\full stack data science and AI\Data Files\Dataset.csv")
Dataset_df
```

1=>

```python
tot_rest=len(Dataset_df)
rest_with_on_delv=Dataset_df['Has Online delivery'].value_counts().get('Yes',0)
per_on_delv=(rest_with_on_delv/tot_rest)*100
print('-------------------------------------------------------------------')
print(f'Percentage of Restaurants that offer Online Delivery:{per_on_delv}')
print('-------------------------------------------------------------------')
```

2=>

```python
avg_rating_with_online=Dataset_df[Dataset_df['Has Online delivery']=='Yes']['Aggregate rating'].mean()
avg_rating_without_online=Dataset_df[Dataset_df['Has Online delivery']=='No']['Aggregate rating'].mean()
print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
print(f'Average rating of restaurant with Online delivery:{avg_rating_with_online}')
print('-------------------------------------------------------------------')
print(f'Average rating of restaurant with Online delivery:{avg_rating_with_online}')
print('-------------------------------------------------------------------')
print('-------------------------------------------------------------------')
```

```python
l=['With Online Delivery','Without Online delivery']
average_rating=[avg_rating_with_online,avg_rating_without_online]
plt.bar(l,average_rating)
plt.xlabel('Online Delivery')
plt.ylabel('average Rating')
plt.title('comparision of Average Ratings')
plt.show()
```
